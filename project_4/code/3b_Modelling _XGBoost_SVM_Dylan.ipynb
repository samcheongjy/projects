```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from datetime import datetime
import calendar
import time
```

# Importing Train

```python
#import merged and clean train dataframe
X_train = pd.read_csv('../data/train_070911.csv')
print(X_train.shape)
X_train.head()
```

```python
pd.set_option('display.max_rows', 20)
pd.set_option('display.max_columns', 83)

#creating target column
y_train = X_train['WnvPresent']

#columns to drop.
drop_columns = ['Date','Trap','month','year_month','carrier','Latitude','Longitude','week_y','WnvPresent','Sunrise','Station','station',
                'Sunset','sunrise_clean','sunset_clean','month','NumMosquitos','year_y','day','WetBulb','Tmin','Tmax','ResultDir']
X_train.drop(columns = drop_columns,inplace=True)

print(X_train.shape,y_train.shape)
X_train.head()
```

# IMPORTING TEST

```python
#import merged and clean test dataframe
X_test = pd.read_csv('../data/train2013.csv')
y_test = X_test['WnvPresent']
X_test.drop(columns = drop_columns,inplace=True)

print(X_test.shape,y_test.shape)
X_test.head()
```

# SCALING

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
X_train = pd.DataFrame(ss.fit_transform(X_train), columns=X_train.columns)
print(X_train.shape)
X_train.describe()
```

```python
X_test = pd.DataFrame(ss.fit_transform(X_test), columns=X_test.columns)
print(X_test.shape)
X_test.describe()
```

# MODELLING - XGBoost

```python
#!pip install xgboost
```

```python
from xgboost import XGBClassifier
```

```python
#to find out the number in the majority and minority class
print(y_train.value_counts(normalize=True))
print(y_train.value_counts())

xgb_weight = y_train.value_counts()[0]/y_train.value_counts()[1]
```

```python
xgb = XGBClassifier(scale_pos_weight=xgb_weight,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, roc_auc_score

def eval_model(y_true, predictions, model=None, X=None):
    cnf_matrix = confusion_matrix(y_true, predictions)

    print('True Negative: ', cnf_matrix[0, 0], '| False Positive: ', cnf_matrix[0, 1])
    print('False Negative: ', cnf_matrix[1, 0], '| True Positive: ', cnf_matrix[1, 1], '\n')

    sensitivity = cnf_matrix[1, 1]/ (cnf_matrix[1, 0] + cnf_matrix[1, 1])
    specificity = cnf_matrix[0, 0]/ (cnf_matrix[0, 1] + cnf_matrix[0, 0])

    print('Sensitivity (TP/ TP + FN): ', sensitivity)
    print('Specificity (TN/ TN + FP): ', specificity, '\n')

    print('Accuracy: ', accuracy_score(y_true, predictions, normalize=True))
    print('Precision: ', precision_score(y_true, predictions))
    if model != None:
        print('Roc-Auc: ', roc_auc_score(y_true, [x[1] for x in model.predict_proba(X)]))
    else:
        pass
    print('\n')
```

# XGBOOST GRIDSEARCH 

```python
from sklearn.model_selection import GridSearchCV
```

```python
start_time = time.time()

param_grid = dict(max_depth = [3,4,5],
                 learning_rate = [0.1,0.3,0.5,0.7,0.9],
                 n_estimators = range(50,200,50)
                 )
grid = GridSearchCV(estimator = xgb,param_grid = param_grid,cv=3,scoring='roc_auc', verbose=1,n_jobs=-1)
grid.fit(X_train, y_train)

best_results = {'params': list(grid.best_params_.items()), 'score': grid.best_score_}

best_model = grid.best_estimator_

best_model = best_model.fit(X_train, y_train)

score = best_model.score(X_test, y_test)

test_predictions = best_model.predict(X_test)

print("{} Score: {:0.3}".format('XGBoost', score.mean().round(3)), '\n')
print('Elapsed Time: {:0.3}'.format( time.time() - start_time), ' seconds', '\n')
print(grid.best_estimator_, '\n')
print('Best Hyperparameters we tested for', '\n', best_results)

eval_model(y_test, test_predictions,model=best_model,X=X_test)
```

```python
feature_import = best_model.feature_importances_
features = X_train.columns
pd.DataFrame(columns=['feature', 'import'], data=list(zip(features, feature_import))).sort_values('import',ascending=False).head(10)
```

# SVM GRIDSEARCH

```python
from imblearn.over_sampling import SMOTE
from sklearn.svm import SVC
svm=SVC(random_state=42,probability=True)
```

```python
#SMOTE-ing to balance the 
smt=SMOTE(random_state=42)
smtXtrain,smtytrain=smt.fit_sample(X_train,y_train)
smtytrain.value_counts()
```

```python
start_time = time.time()

svm_param_grid = dict(gamma = [0.001, 0.01, 0.1, 1],
                C = [0.001, 0.01, 0.1, 1],
                kernel = ['linear', 'poly', 'rbf',]
                 )
grid = GridSearchCV(estimator = svm,param_grid = svm_param_grid,cv=3,scoring='roc_auc', verbose=2,n_jobs=-1)
grid.fit(smtXtrain, smtytrain)

best_results = {'params': list(grid.best_params_.items()), 'score': grid.best_score_}

best_model = grid.best_estimator_

best_model = best_model.fit(smtXtrain, smtytrain)

score = best_model.score(X_test, y_test)

test_predictions = best_model.predict(X_test)

print("{} Score: {:0.3}".format('SVM', score.mean().round(3)), '\n')
print('Elapsed Time: {:0.3}'.format( time.time() - start_time), ' seconds', '\n')
print(grid.best_estimator_, '\n')
print('Best Hyperparameters we tested for', '\n', best_results)

eval_model(y_test, test_predictions,model=best_model,X=X_test)
```

XGBoost had a better ROC-AUC score than SVM.
